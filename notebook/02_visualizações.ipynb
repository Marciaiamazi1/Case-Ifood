```python
import matplotlib.pyplot as plt

# Exemplo de dados (substitua pelos seus dados reais)
pedidos_mes = {
    'Jan': 1200, 'Fev': 1500, 'Mar': 1100,
    'Abr': 1700, 'Mai': 1400, 'Jun': 1800
}

pedidos_semana = {
    'Semana 1': 300, 'Semana 2': 350,
    'Semana 3': 400, 'Semana 4': 380
}

# Configurao do layout
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=False)
fig.suptitle('Pedidos por Ms e por Semana', fontsize=14, weight='bold')

# Grfico de pedidos por ms
meses, valores_mes = zip(*sorted(pedidos_mes.items(), key=lambda x: x[1]))
ax1.barh(meses, valores_mes, color='#4C72B0')
ax1.set_title('Pedidos por Ms')
ax1.set_xlabel('Nmero de Pedidos')
ax1.grid(axis='x', linestyle='--', alpha=0.5)
for i, v in enumerate(valores_mes):
    ax1.text(v + 20, i, str(v), va='center')

# Grfico de pedidos por semana
semanas, valores_sem = zip(*sorted(pedidos_semana.items(), key=lambda x: x[1]))
ax2.barh(semanas, valores_sem, color='#55A868')
ax2.set_title('Pedidos por Semana')
ax2.set_xlabel('Nmero de Pedidos')
ax2.grid(axis='x', linestyle='--', alpha=0.5)
for i, v in enumerate(valores_sem):
    ax2.text(v + 5, i, str(v), va='center')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style="whitegrid")
plt.rcParams['figure.facecolor'] = 'none'

# Dados fictcios (ajuste com seus valores reais)
grupos = ['Controle', 'Cupom']
ticket_medio = [42.30, 43.00]  # ajuste conforme necessrio

plt.figure(figsize=(6, 4))
ax = sns.barplot(x=grupos, y=ticket_medio, palette=['#4C72B0', '#55A868'])
for i, valor in enumerate(ticket_medio):
    ax.text(i, valor + 0.5, f'R$ {valor:.2f}', ha='center', fontweight='bold')
plt.title(' Ticket Mdio por Grupo', fontsize=14)
plt.ylabel('R$')
plt.xlabel('')
plt.tight_layout()
plt.savefig("ticket_medio.png", dpi=300, transparent=True)
plt.show()
```

```python
# Dados de reteno
grupos = ['Controle', 'Cupom']
retencao = [44.1, 46.7]

plt.figure(figsize=(6, 4))
ax = sns.barplot(x=grupos, y=retencao, palette=['#4C72B0', '#55A868'])
for i, valor in enumerate(retencao):
    ax.text(i, valor + 0.5, f'{valor:.1f}%', ha='center', fontweight='bold')
plt.title(' Reteno por Grupo', fontsize=14)
plt.ylabel('% de Reteno')
plt.xlabel('')
plt.tight_layout()
plt.savefig("retencao.png", dpi=300, transparent=True)
plt.show()
```

```python
# Importando as bibliotecas necessrias
import matplotlib.pyplot as plt
import numpy as np

# 1. Grfico para o aumento de converso de 15% no cupom A em relao ao grupo controle
group_control_conversion = 0.20  # Exemplo: 20% de converso no grupo controle
coupon_a_conversion = group_control_conversion * 1.15  # Aumento de 15%

# Plotando o grfico de comparao
labels = ['Grupo Controle', 'Cupom A']
values = [group_control_conversion, coupon_a_conversion]

plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['blue', 'green'])
plt.ylabel('Taxa de Converso')
plt.title('Comparao de Converso: Cupom A vs Grupo Controle')
plt.ylim(0, 1)
plt.show()

# 2. Grfico sobre o comportamento de usurios com alto ticket mdio
ticket_medium = np.random.uniform(0, 100)  # Simulao de ticket mdio
high_ticket = ticket_medium + np.random.uniform(100, 500)  # Simulao de usurios de alto ticket

# Supondo que o cupom A tem melhor desempenho em usurios de alto ticket
user_group = ['Ticket Mdio', 'Alto Ticket']
response_rate = [0.25, 0.40]  # Taxa de resposta hipottica

# Plotando o grfico
plt.figure(figsize=(6, 4))
plt.bar(user_group, response_rate, color=['blue', 'orange'])
plt.ylabel('Taxa de Resposta')
plt.title('Taxa de Resposta ao Cupom A por Grupo de Ticket')
plt.ylim(0, 1)
plt.show()

# 3. Grfico sobre a viabilidade financeira da estratgia no curto prazo
financial_viability = np.random.uniform(0, 50)  # Simulao de custo
financial_return = financial_viability + np.random.uniform(100, 500)  # Retorno esperado

# Plotando o grfico de viabilidade
plt.figure(figsize=(6, 4))
plt.bar(['Custo', 'Retorno'], [financial_viability, financial_return], color=['red', 'green'])
plt.ylabel('Valor (em R$)')
plt.title('Viabilidade Financeira da Estratgia no Curto Prazo')
plt.show()
```
